# Aggregate The  Data
Load the data for each of the years into a single DataFrame, remove any duplicates and then save to disk.

```python
start_year=2001
end_year=2020

# Create empty DataFrame objects
dfTours = pd.DataFrame()
dfMatches = pd.DataFrame()
dfPlayers = pd.DataFrame()
dfRankings = pd.DataFrame()

# Each year add the new data to the DataFrames
for year in range(start_year,end_year):
    newTour    = pd.DataFrame.from_csv(path="./data/dfTours_"+str(year)+".csv", encoding="mbcs", index_col=0)
    dfTours    = pd.concat([dfTours,newTour])

    newMatch   = pd.DataFrame.from_csv(path="./data/dfMatches_"+str(year)+".csv", encoding="mbcs", index_col=[0,1])
    dfMatches  = pd.concat([dfMatches,newMatch])

    newPlayer  = pd.DataFrame.from_csv(path="./data/dfPlayers_"+str(year)+".csv", encoding="mbcs", index_col=0)
    dfPlayers  = pd.concat([dfPlayers,newPlayer])

    newRanking  = pd.DataFrame.from_csv(path="./data/dfRankings_"+str(year)+".csv", encoding="mbcs", index_col=[0,1])
    dfRankings  = pd.concat([dfRankings,newRanking])
    
# Remove the indexes so the next step works correctly
dfTours    = dfTours.reset_index()
dfMatches  = dfMatches.reset_index()
dfPlayers  = dfPlayers.reset_index()
dfRankings = dfRankings.reset_index()

# Remove any duplicate rows
dfTours    = dfTours.drop_duplicates()
dfMatches  = dfMatches.drop_duplicates()
dfPlayers  = dfPlayers.drop_duplicates()
dfRankings = dfRankings.drop_duplicates()

# Put the indexes back
dfTours = dfTours.set_index("TournamentID")
dfMatches  = dfMatches.set_index(["TournamentID","MatchID"])
dfPlayers  = dfPlayers.set_index("PlayerID")
dfRankings = dfRankings.set_index(["PlayerID","MatchDate"])

# A bit of useful debug
print(str(len(dfTours))+" Tournaments")
print(str(len(dfMatches))+" Matches")
print(str(len(dfPlayers))+" Players")
print(str(len(dfRankings))+" Rankings")

# Save the complete DataFrame's
dfTours.to_csv(path_or_buf="./data/dfTours_ALL.csv")
dfMatches.to_csv(path_or_buf="./data/dfMatches_ALL.csv")
dfPlayers.to_csv(path_or_buf="./data/dfPlayers_ALL.csv")
dfRankings.to_csv(path_or_buf="./data/dfRankings_ALL.csv")
```

```python
dfTennis['WinnerWTD'] = 0
dfTennis['WinnerLTD'] = 0
dfTennis['WinnerInWTD'] = 0
dfTennis['WinnerInLTD'] = 0
dfTennis['WinnerOutWTD'] = 0
dfTennis['WinnerOutLTD'] = 0
dfTennis['LoserWTD'] = 0
dfTennis['LoserLTD'] = 0
dfTennis['LoserInWTD'] = 0
dfTennis['LoserInLTD'] = 0
dfTennis['LoserOutWTD'] = 0
dfTennis['LoserOutLTD'] = 0


playerTracker = {}
wWTDCol = list(dfTennis.columns.values).index("WinnerWTD")
wLTDCol = list(dfTennis.columns.values).index("WinnerLTD")
wInWTDCol = list(dfTennis.columns.values).index("WinnerInWTD")
wInLTDCol = list(dfTennis.columns.values).index("WinnerInLTD")
wOutWTDCol = list(dfTennis.columns.values).index("WinnerOutWTD")
wOutLTDCol = list(dfTennis.columns.values).index("WinnerOutLTD")
lWTDCol = list(dfTennis.columns.values).index("LoserWTD")
lLTDCol = list(dfTennis.columns.values).index("LoserLTD")
lInWTDCol = list(dfTennis.columns.values).index("LoserInWTD")
lInLTDCol = list(dfTennis.columns.values).index("LoserInLTD")
lOutWTDCol = list(dfTennis.columns.values).index("LoserOutWTD")
lOutLTDCol = list(dfTennis.columns.values).index("LoserOutLTD")
    
for match_no in range(0,len(dfTennis)):
    #print("Year " + str(dfTennis.iloc[match_no]['Year']) + ", TourNo " + str(dfTennis.iloc[match_no]['TourNo']) + ", MatchNum " + str(dfTennis.iloc[match_no]['MatchNum']))
    #print("Match " + str(dfTennis.index[match_no]) + ", Winner " + dfTennis.iloc[match_no]['WinnerID'] + ", Loser " + dfTennis.iloc[match_no]['LoserID'])
    
    
    match = dfTennis.iloc[match_no]
    playerID = match['WinnerID']

    # Winner
#    (playerID, surface) = list(dfTennis.iloc[match_no]['WinnerID','Surface'])
    if playerID in playerTracker:
        (wWTD, wLTD, wInWTD, wInLTD, wOutWTD, wOutLTD) = playerTracker[playerID]
    else:
        (wWTD, wLTD, wInWTD, wInLTD, wOutWTD, wOutLTD) = (0, 0, 0, 0, 0, 0)

    if playerID == 'vladimir-voltchkov:v211':
        print('Winner: ' + str((wWTD, wLTD, wInWTD, wInLTD, wOutWTD, wOutLTD)) + ' on ' + match['Conditions'])
        
    wWTD = wWTD + 1
    
    if match['Conditions'] == 'Indoor':
        wInWTD = wInWTD + 1
    else:
        wOutWTD = wOutWTD + 1

    if playerID == 'vladimir-voltchkov:v211':
        print('Winner: ' + str((wWTD, wLTD, wInWTD, wInLTD, wOutWTD, wOutLTD)) + ' on ' + match['Conditions'])
    
    playerTracker[playerID] = (wWTD, wLTD, wInWTD, wInLTD, wOutWTD, wOutLTD)
    
    # Loser
    playerID = dfTennis.iloc[match_no]['LoserID']
    if playerID in playerTracker:
        #print("Loser " + str(playerTracker[playerID]))
        (lWTD, lLTD, lInWTD, lInLTD, lOutWTD, lOutLTD) = playerTracker[playerID]
    else:
        (lWTD, lLTD, lInWTD, lInLTD, lOutWTD, lOutLTD) = (0, 0, 0, 0, 0, 0)
 
    if playerID == 'vladimir-voltchkov:v211':
        print('Loser:  ' + str(((lWTD, lLTD, lInWTD, lInLTD, lOutWTD, lOutLTD))) + ' on ' + match['Conditions'] )
        
    lLTD = lLTD + 1
    if match['Conditions'] == 'Indoor':
        lInLTD = lInLTD + 1
    else:
        lOutLTD = lOutLTD + 1
    
    if playerID == 'vladimir-voltchkov:v211':
        print('Loser:  ' + str(((lWTD, lLTD, lInWTD, lInLTD, lOutWTD, lOutLTD))) + ' on ' + match['Conditions'] )
              
    playerTracker[playerID] = (lWTD, lLTD, lInWTD, lInLTD, lOutWTD, lOutLTD)
    
    # print("Wins : " + str(wWTD) + ", Indoor : " + str(wInWTD) + ", Outdoor : " + str(wOutWTD))
    dfTennis.iloc[match_no,wWTDCol] = wWTD
    dfTennis.iloc[match_no,wLTDCol] = wLTD
    dfTennis.iloc[match_no,wInWTDCol] = wInWTD
    dfTennis.iloc[match_no,wInLTDCol] = wInLTD
    dfTennis.iloc[match_no,wOutWTDCol] = wOutWTD
    dfTennis.iloc[match_no,wOutLTDCol] = wOutLTD
    dfTennis.iloc[match_no,lWTDCol] = lWTD
    dfTennis.iloc[match_no,lLTDCol] = lLTD
    dfTennis.iloc[match_no,lInWTDCol] = lInWTD
    dfTennis.iloc[match_no,lInLTDCol] = lInLTD
    dfTennis.iloc[match_no,lOutWTDCol] = lOutWTD
    dfTennis.iloc[match_no,lOutLTDCol] = lOutLTD

    #print("After " + str((wWTD, wLTD, lWTD, lLTD)))
    #if match_no > 20:
    #    break
```
