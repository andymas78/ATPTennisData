```python
dfTennis = dfTours
dfTennis = dfTennis.reset_index()

dfMatches = dfMatches.reset_index()
dfPlayers = dfPlayers.reset_index()

dfTennis = dfTennis.merge(dfMatches, how='inner', on='TournamentID', suffixes=['l','r'])
dfTennis = dfTennis.merge(dfPlayers, how='inner', left_on='WinnerID', right_on='PlayerID', suffixes=[None, '_Winner'])
dfTennis = dfTennis.merge(dfPlayers, how='inner', left_on='LoserID', right_on='PlayerID', suffixes=['_Winner', '_Loser'])

# dfTennis['MatchOrder'] = dfTennis['Year'] * 100000 + dfTennis['TourNo'] * 100 + dfTennis['MatchNum']
#dfTennis['MatchIDX'] = dfTennis['TournamentID'] + dfTennis['MatchID'] + str(dfTennis['MatchNum'])

#dfTennis['WinsToDate'] = list(map(lambda x : len(dfTennis[(dfTennis['WinnerID'] == dfTennis.iloc[x]['WinnerID']) & 
#                                                          (dfTennis['MatchIDX'] < dfTennis.iloc[x]['MatchIDX'])]), dfTennis.index.values))

dfTennis = dfTennis.set_index(['Year', 'TourNo', 'MatchNum'])
dfTennis = dfTennis.sort_index()
dfTennis.columns
```

```python
dfTennis['WinnerWTD'] = 0
dfTennis['WinnerLTD'] = 0
dfTennis['LoserWTD'] = 0
dfTennis['LoserLTD'] = 0

playerTracker = {}
wWTDCol = list(dfTennis.columns.values).index("WinnerWTD")
wLTDCol = list(dfTennis.columns.values).index("WinnerLTD")
lWTDCol = list(dfTennis.columns.values).index("LoserWTD")
lLTDCol = list(dfTennis.columns.values).index("LoserLTD")
    
for match_no in range(0,len(dfTennis)):
    #print("Year " + str(dfTennis.iloc[match_no]['Year']) + ", TourNo " + str(dfTennis.iloc[match_no]['TourNo']) + ", MatchNum " + str(dfTennis.iloc[match_no]['MatchNum']))
    #print("Match " + str(dfTennis.index[match_no]) + ", Winner " + dfTennis.iloc[match_no]['WinnerID'] + ", Loser " + dfTennis.iloc[match_no]['LoserID'])
    
    

    
    # Winner
    playerID = dfTennis.iloc[match_no]['WinnerID']
    if playerID in playerTracker:
        #print("Winner " + str(playerTracker[playerID]))
        (wWTD, wLTD) = playerTracker[playerID]
    else:
        (wWTD, wLTD) = (0, 0)
        
    wWTD = wWTD + 1
    playerTracker[playerID] = (wWTD, wLTD)
    
    # Loser
    playerID = dfTennis.iloc[match_no]['LoserID']
    if playerID in playerTracker:
        #print("Loser " + str(playerTracker[playerID]))
        (lWTD, lLTD) = playerTracker[playerID]
    else:
        (lWTD, lLTD) = (0, 0)
        
    lLTD = lLTD + 1
    
    playerTracker[playerID] = (lWTD, lLTD)
    
    
    dfTennis.iloc[match_no,wWTDCol] = wWTD
    dfTennis.iloc[match_no,wLTDCol] = wLTD
    dfTennis.iloc[match_no,lWTDCol] = lWTD
    dfTennis.iloc[match_no,lLTDCol] = lLTD

    #print("After " + str((wWTD, wLTD, lWTD, lLTD)))
    #if match_no > 20:
    #    break
```
